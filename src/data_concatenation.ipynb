```python

def convertime(data):
    data['month'] = 0
    data['year'] = 0
    for index, row in data.iterrows():
        year = (row[3])[:4]
        month = (row[3])[5:7]
        data.set_value(index, 'year', year)
        data.set_value(index, 'month', month)
    return data

pressure.drop(3)
pressure = convertime(pressure)

pressure.info()
pressure.head()
```
