```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from dateutil.parser import parse
from getState import getState
```

```python
baro_data = pd.read_csv('clean_datasets/baro_data.csv', index_col=0)
air_temp_data = pd.read_csv('clean_datasets/air_temp_data.csv', index_col=0)
humidity_data = pd.read_csv('clean_datasets/humidity_data.csv', index_col=0)
```

```python
def convert_time(data):
    data['month'] = 0
    data['year'] = 0
    for index, row in data.iterrows():
        year = (row[2])[:4]
        month = (row[2])[5:7]
        data.set_value(index, 'year', year)
        data.set_value(index, 'month', month)
    return data
```

```python
#making month and year column
baro_data_t = convert_time(baro_data).drop([' measured_on'], axis=1)
air_temp_data_t = convert_time(air_temp_data).drop(['measured_on'], axis=1)
humidity_data_t = convert_time(humidity_data).drop([' measured_on'], axis=1)
```

```python
def get_state(data):
    data['state'] = 'null'
    coordToState = {}
    for index, row_2 in data.iterrows():
        lat = row_2[0]
        lon = row_2[1]
        if (lat, lon) in coordToState:
            state = coordToState[(lat, lon)]
        else:
            state = getState(lat, lon)
            coordToState[(lat, lon)] = state
        data.set_value(index, 'state', state)
```

```python
air_temp_sample = air_temp_data_t.sample(n=20000)
baro_data_sample = baro_data_t.sample(n=20000)
humidity_sample = humidity_data_t.sample(n=20000)
```
